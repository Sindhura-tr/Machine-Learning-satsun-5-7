## Association Rule Mining

```python
import pandas as pd
df= pd.read_csv('/content/store_data.csv',header=None)
df.head()
```

```python
df.shape
```

processing data for ARM algorithm

```python
records = []
for i in range(df.shape[0]):
  records.append([str(df.values[i,j]) for j in range(df.shape[1])])
```

```python
print(records[:5])
```

```python
conf = 0.2
sup = (7*5)/len(df)
lift = 3
```

```python
from apyori import apriori
association_rules = apriori(records,min_support=sup,min_confidence=conf,min_lift=lift)
association_rules = list(association_rules)
```

```python
print(len(association_rules))
```

```python
association_rules[0]
```

```python
association_rules[1]
```

```python
association_rules[1][0]
```

```python
for item in association_rules:
  pair = item[0]

  items = [x for x in pair]
  print("Rule: " + items[0] + "-- >" + items[1])

  support = item[1]
  print("Support" + str(item[1]))

  confi = str(item[2][0][2])
  lift = str(item[2][0][3])
  print("Confidence" + confi)
  print("Lift" + lift)
  print("==========================")
```

```python
rule = []
supports = []
confis = []
lifts = []

for item in association_rules:
  pair = item[0]

  items = [x for x in pair]
  r = items[0] + "-- >" + items[1]
  rule.append(r)

  support = item[1]
  supports.append(support)
  #print("Support" + str(item[1]))

  confi = item[2][0][2]
  lift = item[2][0][3]
  confis.append(confi)
  lifts.append(lift)
```

```python
rule
```

```python
lifts
```

## Create dictionary for dataframe

```python
dct = {
    "Rule": rule,
    "Support":supports,
    "Confidence":confis,
    "Lift":lifts
}
```

```python
df_association = pd.DataFrame(dct)
df_association.head()
```

## SAve the association results to csv file

```python
df_association.to_csv("Association_results.csv",index=False)
```
